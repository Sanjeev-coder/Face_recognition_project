### working with images with matplotlib

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# read an image
img = cv2.imread("sanjeev.png")
# we can use matplotlib to show this image
plt.imshow(img)
# matplot libe reads images r->g->b as b->g->r that's why it will show picture of different color, if we use cv2.imshow(img) then it will read bgr as bgr 
```

```python
# here we will chnage the manner to show picture by using a function which opencv provide
newImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # this conveting from bgr to rgb
plt.imshow(newImg)
plt.show()  
```

```python
cv2.imshow("my",img)
cv2.waitKey(25)   # zero means infinite amount time, if we say (25), the window will wait for 25ms. 
cv2.destroyAllWindows()  # we are destroying window after seen
```
